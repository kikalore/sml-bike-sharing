```python
import pandas as pd
import requests
file_path = "bike-historical-data/202501-e-motion-trento-stats.parquet"
```

```python
# Read historical data
df = pd.read_parquet(file_path, engine='pyarrow')

print(df.head())
print(df.columns)
print(df.info())
```

```python
# Get real-time data
network_id = "e-motion-trento" 
url = f"https://api.citybik.es/v2/networks/{network_id}"

resp = requests.get(url).json()
```

```python
# Stations info
stations = resp['network']['stations']

# print first 5 stations
for s in stations[:5]:
    print(f"{s['name']} -> Bikes: {s['free_bikes']}, Empty slots: {s['empty_slots']}")
```
